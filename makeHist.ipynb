## Part 3

```python
num = 0.3
file1 = open('pt_3/beta_{}.txt'.format(num), 'r')
#file1 = open('pt_2/beta_max.txt', 'r')
line = file1.readline().rstrip()
snapshot_ar = np.empty((81,81), dtype=int)
x = 0
for i in line:
    snapshot_ar[int(x/81), (x%81)] = ord(i)-48
    x += 1
print(snapshot_ar)
```

```python
# Convert config into snapshot
def convertConfToSnap2(snapshot_ar):
    pylab.matshow(snapshot_ar)
    return snapshot_ar
first_ar = convertConfToSnap2(snapshot_ar)
```

```python
def convert_to_27(first_ar):
    second_ar = np.empty((27,27))
    for x in range(0,81):
        for y in range(0,81):
            second_ar[int(x/3),int(y/3)] = np.round(np.mean(first_ar[x:x+3,y:y+3]))
            y += 3
        x += 3
    return second_ar
second_ar = convert_to_27(first_ar)
pylab.matshow(second_ar)
```

```python
def convert_to_9(second_ar):
    third_ar = np.empty((9,9))
    for x in range(0,27):
        for y in range(0,27):
            third_ar[int(x/3),int(y/3)] = np.round(np.mean(second_ar[x:x+3,y:y+3]))
            y += 3
        x += 3
    return third_ar
third_ar = convert_to_9(second_ar)
pylab.matshow(third_ar)
```
